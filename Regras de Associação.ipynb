```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
dados = pd.read_csv('Mobile Apps(Full).csv')
dados_cv = dados.iloc[:,0:15].values
te = TransactionEncoder()
arr = te.fit(dados_cv).transform(dados_cv)
df = pd.DataFrame(arr,columns=te.columns_)
df.to_csv('/home/wellington/Documentos/Anaconda/pesquisa/df.csv')
```

```python
dados = pd.read_csv('df.csv')
dados
```

```python
suportes = apriori(dados, min_support=0.2, use_colnames=True)
suportes
```

```python
regras = association_rules(suportes, metric="confidence",  min_threshold=0.8)
regras
```

```python
regras.to_csv(r'/home/wellington/Documentos/Anaconda/pesquisa/resultado.csv')
```
